```python
# Merging and Joining DataFrames
df3 = pd.DataFrame({'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'], 
'State': ['NY', 'CA', 'IL', 'TX', 'AZ']})
df_merged = pd.merge(df2, df3, on='City', how='inner') # inner join
df_merged.head(20) # first 5 rows of merged dataframe
# df_merged.info()
```

```python
df_merged = pd.merge(df2, df3, on='City', how='outer') # outer join
df_merged.head(20) # first 5 rows of merged dataframe
```

```python
df_merged = pd.merge(df2, df3, on='City', how='left') # left join
df_merged.head(20) # first 5 rows of merged dataframe
```

```python
df_merged = pd.merge(df2, df3, on='City', how='right') # right join
df_merged.head(20) # first 5 rows of merged dataframe
```

```python
# merge and join in pandas are similar but not exactly the same:

# Merge
# 1. Works like SQL joins.
# 2. Can join on one or more columns with different names.
# 3. Flexible: supports on, left_on, right_on, how (inner, left, right, outer).
# 4. Syntax: pd.merge(df1, df2, on='key', how='inner')

# Join
# 1. Simpler, mainly for joining on the index (row labels) or a column.
# 2. By default, joins on the index of the calling DataFrame.
# 3. Less flexible for joining on columns with different names.
# Syntax: df1.join(df2, how='left')

# Summary:

# Use merge for SQL-style joins on columns.
# Use join for index-based joins or quick merges on the index.
```
